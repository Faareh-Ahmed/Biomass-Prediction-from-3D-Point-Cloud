# Data Pre-Processing

```python
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from torch.utils.data import DataLoader, Dataset

# Data preprocessing for point cloud
def preprocess_point_cloud(point_cloud):
    # Convert point cloud to numerical values (ignore headers, nan values, and ensure X, Y, Z are floats)
    point_cloud = np.array(point_cloud[2:], dtype=np.float32)  # Skip the header row
    # Remove rows where any NaN exists (or handle it as needed)
    point_cloud = point_cloud[~np.isnan(point_cloud).any(axis=1)]
    return point_cloud

from torch.nn.functional import pad

def pad_point_cloud(point_cloud, max_points=1024):
    num_points = point_cloud.shape[0]

    # If the number of points is less than the max, pad with zeros
    if num_points < max_points:
        padding = max_points - num_points
        point_cloud = np.pad(point_cloud, ((0, padding), (0, 0)), mode='constant', constant_values=0)
    # If the number of points is greater than the max, truncate
    elif num_points > max_points:
        point_cloud = point_cloud[:max_points, :]

    return point_cloud

# Pad point clouds in the dataset
max_points = 1024  # Define the max number of points

def load_data_with_padding(df, max_points=1024):
    point_clouds = []
    biomass_values = []

    for _, row in df.iterrows():
        point_cloud = preprocess_point_cloud(row['point_cloud'])
        point_cloud = pad_point_cloud(point_cloud, max_points)
        point_clouds.append(point_cloud)
        biomass_values.append(row['biomass'])

    return point_clouds, biomass_values

# Load point clouds and biomass data
point_clouds, biomass_values = load_data_with_padding(df_point_clouds)

# Normalize the point cloud data
scaler = StandardScaler()
point_clouds = [scaler.fit_transform(pc) for pc in point_clouds]  # Apply normalization to each point cloud


# Train-test split
X_train, X_test, y_train, y_test = train_test_split(point_clouds, biomass_values, test_size=0.2, random_state=42)

# Create a custom Dataset
class BiomassDataset(Dataset):
    def __init__(self, point_clouds, biomass_values):
        self.point_clouds = point_clouds
        self.biomass_values = biomass_values

    def __len__(self):
        return len(self.point_clouds)

    def __getitem__(self, idx):
        return torch.tensor(self.point_clouds[idx][:, :3], dtype=torch.float32), torch.tensor(self.biomass_values[idx], dtype=torch.float32)

# Convert to DataLoader
train_dataset = BiomassDataset(X_train, y_train)
test_dataset = BiomassDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)
```

# Self Attention based PointNet Regression

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class SelfAttention(nn.Module):
    def __init__(self, dim, heads=4):
        super(SelfAttention, self).__init__()
        self.heads = heads
        self.scale = (dim // heads) ** -0.5  # Scaling factor

        self.qkv = nn.Linear(dim, dim * 3, bias=False)  # Create Q, K, V
        self.proj = nn.Linear(dim, dim)  # Final projection

    def forward(self, x):
        B, N, C = x.shape  # Batch, Num_Points, Channels
        qkv = self.qkv(x).chunk(3, dim=-1)  # Split into Q, K, V

        # Reshape for multi-head attention
        q, k, v = [t.view(B, N, self.heads, C // self.heads).transpose(1, 2) for t in qkv]

        attn = (q @ k.transpose(-2, -1)) * self.scale  # Compute attention scores
        attn = attn.softmax(dim=-1)  # Normalize

        x = (attn @ v).transpose(1, 2).contiguous().view(B, N, C)  # Apply attention and reshape
        x = self.proj(x)  # Final projection

        return x

class PointNetRegressor(nn.Module):
    def __init__(self):
        super(PointNetRegressor, self).__init__()

        # Shared MLP layers (Feature Extractors)
        self.conv1 = nn.Conv1d(3, 64, 1)
        self.bn1 = nn.BatchNorm1d(64)
        self.conv2 = nn.Conv1d(64, 128, 1)
        self.bn2 = nn.BatchNorm1d(128)
        self.conv3 = nn.Conv1d(128, 256, 1)
        self.bn3 = nn.BatchNorm1d(256)

        # Self-Attention Layer (After feature extraction)
        self.self_attention = SelfAttention(dim=256, heads=4)

        # Global feature extraction
        self.fc1 = nn.Linear(256, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 1)

        # Dropout layer
        self.dropout = nn.Dropout(0.3)

        # Weight initialization
        self.apply(self.initialize_weights)

    def initialize_weights(self, m):
        if isinstance(m, nn.Conv1d) or isinstance(m, nn.Linear):
            nn.init.xavier_normal_(m.weight)
            if m.bias is not None:
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = x.transpose(1, 2)  # Change to (B, Channels, Points)

        # Feature Extraction (Convolutions + GELU)
        x = F.gelu(self.bn1(self.conv1(x)))
        x = F.gelu(self.bn2(self.conv2(x)))
        x = F.gelu(self.bn3(self.conv3(x)))

        x = x.transpose(1, 2)  # (B, Points, Features) for Attention

        # Apply Self-Attention
        x = self.self_attention(x)

        # Global Average Pooling
        x = torch.mean(x, 1)  # Average over all points

        # Fully Connected Layers
        x = F.gelu(self.fc1(x))
        x = self.dropout(x)
        x = F.gelu(self.fc2(x))
        x = self.fc3(x)  # Output regression value

        return x

# Initialize the model
model = PointNetRegressor()

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)  # L2 Regularization

# Learning rate scheduler
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.7)
```

```python
import torch
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluate_model(model, val_loader):
    model.eval()
    predictions = []
    biomass_values = []

    with torch.no_grad():
        for data in val_loader:
            point_clouds, biomass = data
            outputs = model(point_clouds)

            # Squeeze and convert to numpy
            outputs = outputs.squeeze().cpu().numpy()

            # Check if it's a single prediction or a batch
            if outputs.ndim == 0:  # Single prediction
                predictions.append(outputs.item())
            else:  # Batch of predictions
                predictions.extend(outputs.tolist())  # Use tolist() to convert numpy array to list

            biomass_values.extend(biomass.cpu().numpy().tolist()) # Convert to list to use extend

    predictions = np.array(predictions)
    biomass_values = np.array(biomass_values)

    mse = mean_squared_error(biomass_values, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(biomass_values, predictions)
    r2 = r2_score(biomass_values, predictions)

    print(f'Mean Squared Error (MSE): {mse:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
    print(f'Mean Absolute Error (MAE): {mae:.4f}')
    print(f'R² Score: {r2:.4f}')
```

# BIONET Architecture

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim


class CompletionModule(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=1024, output_dim=3):
        super(CompletionModule, self).__init__()
        self.hidden_dim = hidden_dim  # Store hidden_dim as an attribute
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.ReLU(),
            nn.Linear(512, hidden_dim),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dim, 512),
            nn.ReLU(),
            nn.Linear(512, output_dim)
        )

    def forward(self, x):
        batch_size, num_points, _ = x.shape
        x_flat = x.view(batch_size * num_points, -1)  # Flatten to (batch_size * num_points, input_dim)
        encoded_features = self.encoder(x_flat)  # (batch_size * num_points, hidden_dim)
        completed_points_flat = self.decoder(encoded_features)  # (batch_size * num_points, output_dim)
        completed_points = completed_points_flat.view(batch_size, num_points, -1)  # Reshape to (batch_size, num_points, output_dim)
        encoded_features = encoded_features.view(batch_size, num_points, self.hidden_dim)  # Reshape to (batch_size, num_points, hidden_dim)
        return completed_points, encoded_features
```

```python
class RegularizationModule(nn.Module):
    def __init__(self, feature_dim=1024):
        super(RegularizationModule, self).__init__()
        self.mlp_query = nn.Sequential(
            nn.Linear(feature_dim, feature_dim),
            nn.ReLU()
        )
        self.mlp_key = nn.Sequential(
            nn.Linear(feature_dim, feature_dim),
            nn.ReLU()
        )
        self.mlp_value = nn.Sequential(
            nn.Linear(feature_dim, feature_dim),
            nn.ReLU()
        )

    def forward(self, reference_features, plot_features):
        # reference_features: (batch_size, feature_dim)
        # plot_features: (batch_size, num_points, feature_dim)
        query = self.mlp_query(reference_features)
        key = self.mlp_key(plot_features)
        value = self.mlp_value(plot_features)
        attention_scores = torch.matmul(query.unsqueeze(1), key.transpose(1, 2))
        attention_scores = F.softmax(attention_scores, dim=-1)
        regularized_features = torch.matmul(attention_scores, value).squeeze(1)
        return regularized_features
```

```python
class ProjectionModule(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=1024, grid_size=32):
        super(ProjectionModule, self).__init__()
        self.grid_size = grid_size
        self.cnn = nn.Sequential(
            nn.Conv2d(2, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Flatten(),
            nn.Linear(64 * (grid_size//4) * (grid_size//4), hidden_dim),
            nn.ReLU()
        )

    def forward(self, x):
        x_2d = x[:, :, :2]
        batch_size = x_2d.size(0)
        x_2d = x_2d.view(batch_size, self.grid_size, self.grid_size, 2)
        x_2d = x_2d.permute(0, 3, 1, 2)
        features = self.cnn(x_2d)
        return features
```

```python
class FusionBlock(nn.Module):
    def __init__(self, feature_dim=1024):
        super(FusionBlock, self).__init__()
        self.mlp = nn.Sequential(
            nn.Linear(feature_dim * 3, feature_dim),
            nn.ReLU(),
            nn.Linear(feature_dim, 3)  # Output 3 weights
        )

    def forward(self, v_c, v_r, v_p):
        combined = torch.cat([v_c, v_r, v_p], dim=1)
        weights = F.softmax(self.mlp(combined), dim=1)
        fused = weights[:, 0:1] * v_c + weights[:, 1:2] * v_r + weights[:, 2:3] * v_p
        return fused
```

```python
class BioNet(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=1024, grid_size=32):
        super(BioNet, self).__init__()
        self.completion_module = CompletionModule(input_dim, hidden_dim)
        self.regularization_module = RegularizationModule(hidden_dim)
        self.projection_module = ProjectionModule(input_dim, hidden_dim, grid_size)
        self.fusion_block = FusionBlock(hidden_dim)
        self.final_mlp = nn.Sequential(
            nn.Linear(hidden_dim, 512),
            nn.ReLU(),
            nn.Linear(512, 1)
        )

    def forward(self, x):
        completed_points, encoded_features = self.completion_module(x)
        reference_features = encoded_features.mean(dim=1)  # Global feature
        v_c = reference_features
        v_r = self.regularization_module(reference_features, encoded_features)
        v_p = self.projection_module(completed_points)
        fused_features = self.fusion_block(v_c, v_r, v_p)
        biomass_prediction = self.final_mlp(fused_features)
        return biomass_prediction
```

```python
# Initialize the model, optimizer, and loss function
model = BioNet()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
criterion = nn.SmoothL1Loss()  # As mentioned in the paper

# Dynamic Learning Rate Scheduler
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',       # Reduce LR when validation loss stops decreasing
    factor=0.5,       # Reduce the learning rate by a factor of 0.5
    patience=3,       # Wait 5 epochs before reducing LR after no improvement
    verbose=True      # Print LR changes for monitoring
)

# Training function
def train(model, train_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0
    for batch_idx, (point_clouds, biomass_values) in enumerate(train_loader):
        point_clouds, biomass_values = point_clouds.to(device), biomass_values.to(device)

        # Forward pass
        optimizer.zero_grad()
        predicted_biomass = model(point_clouds)
        loss = criterion(predicted_biomass, biomass_values.unsqueeze(1))  # Add a dimension to match predictions

        # Backward pass
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(train_loader)

# Evaluation function
def evaluate(model, test_loader, criterion, device):
    model.eval()
    total_loss = 0.0
    with torch.no_grad():
        for batch_idx, (point_clouds, biomass_values) in enumerate(test_loader):
            point_clouds, biomass_values = point_clouds.to(device), biomass_values.to(device)

            # Forward pass
            predicted_biomass = model(point_clouds)
            loss = criterion(predicted_biomass, biomass_values.unsqueeze(1))

            total_loss += loss.item()

    return total_loss / len(test_loader)

# Training loop
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

num_epochs = 50
for epoch in range(num_epochs):
    train_loss = train(model, train_loader, optimizer, criterion, device)
    test_loss = evaluate(model, test_loader, criterion, device)
    # Adjust the learning rate based on validation loss
    scheduler.step(test_loss)

    # Print the current learning rate
    current_lr = optimizer.param_groups[0]['lr']
    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, LR: {current_lr:.6f}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np

# Define the PointNet-based regression model
class PointNetRegression(nn.Module):
    def __init__(self, input_dim=4, output_dim=1):
        super(PointNetRegression, self).__init__()



        # Input layer for point features (e.g., XYZ coordinates, intensity)
        self.fc1 = nn.Linear(input_dim, 64)
        self.bn1 = nn.BatchNorm1d(64)

        # Hidden layers
        self.fc2 = nn.Linear(64, 128)
        self.bn2 = nn.BatchNorm1d(128)

        self.fc3 = nn.Linear(128, 1024)
        self.bn3 = nn.BatchNorm1d(1024)

        # Global pooling layer (global sum pooling for regression)
        self.global_pool = nn.Identity()

        # Fully connected layers for the regression output
        self.fc4 = nn.Linear(1024, 512)
        self.bn4 = nn.BatchNorm1d(512)

        self.fc5 = nn.Linear(512, 256)
        self.bn5 = nn.BatchNorm1d(256)

        self.fc6 = nn.Linear(256, output_dim)  # Output AGB/wood volume

        # Activation function: GELU
        self.gelu = nn.GELU()

    def forward(self, x, mask=None):

        # print(f"Shape before applying mask: {x.shape}")  # Print the shape before masking

        # Input x has shape (batch_size, num_points, input_dim)
        batch_size, num_points, input_dim = x.size()

         # Apply MLP to each point individually (point-wise feature extraction)
        x = self.fc1(x)  # Shape: (batch_size, num_points, 64)
        x = self.gelu(self.bn1(x.transpose(1, 2)))  # Permute to (batch_size, 64, num_points) for batch norm
        x = self.fc2(x.transpose(1, 2))  # Transpose back to (batch_size, num_points, 128)
        x = self.gelu(self.bn2(x.transpose(1, 2)))  # Batch norm applied to (batch_size, 128, num_points)

        x = self.fc3(x.transpose(1, 2))  # Continue with the same process
        x = self.gelu(self.bn3(x.transpose(1, 2)))

        # print(f"Shape of x before applying mask: {x.shape}")  # Print the shape before masking
        # print(f"Shape of mask before applying mask: {mask.shape}")  # Print the shape before masking


        # Apply mask to ignore padded points (if provided)
        if mask is not None:
            mask = mask.unsqueeze(1)  # Shape: (batch_size, 1, num_points)
            x = x * mask  # Shape: (batch_size, num_points, 1024)

        # Global sum pooling: sum all point features for each sample in the batch
        x = torch.sum(x, dim=2)  # Shape: (batch_size, 1024)

        # Fully connected layers for regression
        x = self.gelu(self.bn4(self.fc4(x)))
        x = self.gelu(self.bn5(self.fc5(x)))
        x = self.fc6(x)  # Output is the regression value

        return x
```

```python
# Preprocess point clouds: remove headers and handle NaNs
# Preprocess point clouds: remove headers, center the data, and apply scaling
def preprocess_point_clouds_with_centering(point_clouds):
    processed_point_clouds = []

    for pc in point_clouds:
        # Remove the header (first two rows)
        pc_cleaned = pc[2:]  # Skip the first row (['//X', 'Y', 'Z', 'intensity'])

        # Convert data to float and replace NaNs with 0
        pc_cleaned = pc_cleaned.astype(np.float32)

        # Extract the X, Y, Z, and intensity columns
        X = pc_cleaned[:, 0]
        Y = pc_cleaned[:, 1]
        Z = pc_cleaned[:, 2]
        intensity = pc_cleaned[:, 3]

        # Center the X, Y, Z coordinates by subtracting the mean
        X_centered = X - np.mean(X)
        Y_centered = Y - np.mean(Y)
        Z_centered = Z - np.mean(Z)

        # Apply scaling: divide X, Y by 15 and Z by 20
        X_scaled = X_centered / 15.0
        Y_scaled = Y_centered / 15.0
        Z_scaled = Z_centered / 20.0

        # Stack the processed coordinates and intensity back into a point cloud
        pc_processed = np.stack([X_scaled, Y_scaled, Z_scaled, intensity], axis=1)

        processed_point_clouds.append(pc_processed)

    return processed_point_clouds

# Function to pad point clouds to have the same number of points
def pad_point_clouds(point_clouds, max_num_points):
    padded_point_clouds = []
    masks = []

    for pc in point_clouds:
        num_points = pc.shape[0]
        if num_points < max_num_points:
            # Pad with zeros
            pad_size = max_num_points - num_points
            padded_pc = np.pad(pc, ((0, pad_size), (0, 0)), 'constant', constant_values=0)  # Pad to the max_num_points
            mask = np.pad(np.ones(num_points), (0, pad_size), 'constant', constant_values=0)  # Create a mask
        else:
            padded_pc = pc[:max_num_points]  # Truncate if necessary
            mask = np.ones(max_num_points)  # Full mask if no padding is needed

        padded_point_clouds.append(padded_pc)
        masks.append(mask)

    return np.array(padded_point_clouds), np.array(masks)
```
