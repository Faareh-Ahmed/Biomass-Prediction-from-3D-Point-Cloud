```python
import torch
from torch.utils.data import Dataset

class PointCloudDataset(Dataset):
    def __init__(self, point_clouds, biomass_values):
        self.point_clouds = point_clouds
        self.biomass_values = biomass_values

    def __len__(self):
        return len(self.biomass_values)

    def __getitem__(self, idx):
        point_cloud = self.point_clouds[idx]
        biomass = self.biomass_values[idx]
        return torch.tensor(point_cloud, dtype=torch.float32), torch.tensor(biomass, dtype=torch.float32)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class KPConvLayer(nn.Module):
    def __init__(self, input_dim, output_dim, kernel_size=15, num_kernel_points=8):
        """
        A simplified KPConv layer.

        Args:
            input_dim (int): The dimension of the input point features.
            output_dim (int): The dimension of the output point features.
            kernel_size (int): The size of the convolution kernel.
            num_kernel_points (int): The number of kernel points to use in convolution.
        """
        super(KPConvLayer, self).__init__()

        self.input_dim = input_dim
        self.output_dim = output_dim
        self.kernel_size = kernel_size
        self.num_kernel_points = num_kernel_points

        # Initialize the weights for the kernel points
        self.kernel_weights = nn.Parameter(torch.randn(num_kernel_points, input_dim, output_dim))

        # Learnable positions of the kernel points
        self.kernel_points = nn.Parameter(torch.randn(num_kernel_points, 3))  # Assume 3D positions for the kernel points

        # A small MLP to map the distances between the input points and the kernel points
        self.mlp = nn.Sequential(
            nn.Linear(3, 16),
            nn.ReLU(),
            nn.Linear(16, output_dim)
        )

    def forward(self, point_features, point_positions):
        """
        Forward pass of the KPConv layer.

        Args:
            point_features (Tensor): The features of the input points (batch_size, num_points, input_dim).
            point_positions (Tensor): The positions of the input points (batch_size, num_points, 3).

        Returns:
            Tensor: The convolved features of the points.
        """
        # Compute the distances between input points and kernel points
        distances = torch.cdist(point_positions, self.kernel_points)

        # Use the distances to calculate the weight of each kernel point for each input point
        weights = torch.exp(-distances)  # You can adjust this weighting scheme

        # Apply the MLP to the distances
        weights = self.mlp(distances)

        # Apply the kernel weights to the input features
        point_features = point_features.unsqueeze(2)  # Add an extra dimension for kernel points
        output_features = torch.matmul(weights.unsqueeze(3), self.kernel_weights)  # Shape: (batch_size, num_points, num_kernel_points, output_dim)
        output_features = torch.sum(output_features, dim=2)  # Sum over the kernel points

        return output_features
```

```python
class KPConvModel(nn.Module):
    def __init__(self, num_points):
        super(KPConvModel, self).__init__()

        # Define KPConv layers
        self.kpconv1 = KPConvLayer(input_dim=4, output_dim=64)
        self.kpconv2 = KPConvLayer(input_dim=64, output_dim=128)
        self.kpconv3 = KPConvLayer(input_dim=128, output_dim=256)

        # Define a fully connected layer for the final regression output
        self.fc = nn.Sequential(
            nn.Linear(256 * num_points, 512),
            nn.ReLU(),
            nn.Linear(512, 1)  # Output is a single value (biomass)
        )

    def forward(self, x, positions):
        # Forward pass through KPConv layers
        x = self.kpconv1(x, positions)
        x = self.kpconv2(x, positions)
        x = self.kpconv3(x, positions)

        # Flatten for the fully connected layer
        x = x.view(x.size(0), -1)

        # Pass through fully connected layers
        x = self.fc(x)

        return x
```

# PointNet

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np

# Define the PointNet-based regression model
class PointNetRegression(nn.Module):
    def __init__(self, input_dim=4, output_dim=1):
        super(PointNetRegression, self).__init__()



        # Input layer for point features (e.g., XYZ coordinates, intensity)
        self.fc1 = nn.Linear(input_dim, 64)
        self.bn1 = nn.BatchNorm1d(64)

        # Hidden layers
        self.fc2 = nn.Linear(64, 128)
        self.bn2 = nn.BatchNorm1d(128)

        self.fc3 = nn.Linear(128, 1024)
        self.bn3 = nn.BatchNorm1d(1024)

        # Global pooling layer (global sum pooling for regression)
        self.global_pool = nn.Identity()

        # Fully connected layers for the regression output
        self.fc4 = nn.Linear(1024, 512)
        self.bn4 = nn.BatchNorm1d(512)

        self.fc5 = nn.Linear(512, 256)
        self.bn5 = nn.BatchNorm1d(256)

        self.fc6 = nn.Linear(256, output_dim)  # Output AGB/wood volume

        # Activation function: GELU
        self.gelu = nn.GELU()

    def forward(self, x, mask=None):

        # print(f"Shape before applying mask: {x.shape}")  # Print the shape before masking

        # Input x has shape (batch_size, num_points, input_dim)
        batch_size, num_points, input_dim = x.size()

         # Apply MLP to each point individually (point-wise feature extraction)
        x = self.fc1(x)  # Shape: (batch_size, num_points, 64)
        x = self.gelu(self.bn1(x.transpose(1, 2)))  # Permute to (batch_size, 64, num_points) for batch norm
        x = self.fc2(x.transpose(1, 2))  # Transpose back to (batch_size, num_points, 128)
        x = self.gelu(self.bn2(x.transpose(1, 2)))  # Batch norm applied to (batch_size, 128, num_points)

        x = self.fc3(x.transpose(1, 2))  # Continue with the same process
        x = self.gelu(self.bn3(x.transpose(1, 2)))

        # print(f"Shape of x before applying mask: {x.shape}")  # Print the shape before masking
        # print(f"Shape of mask before applying mask: {mask.shape}")  # Print the shape before masking


        # Apply mask to ignore padded points (if provided)
        if mask is not None:
            mask = mask.unsqueeze(1)  # Shape: (batch_size, 1, num_points)
            x = x * mask  # Shape: (batch_size, num_points, 1024)

        # Global sum pooling: sum all point features for each sample in the batch
        x = torch.sum(x, dim=2)  # Shape: (batch_size, 1024)

        # Fully connected layers for regression
        x = self.gelu(self.bn4(self.fc4(x)))
        x = self.gelu(self.bn5(self.fc5(x)))
        x = self.fc6(x)  # Output is the regression value

        return x
```

```python
# Preprocess point clouds: remove headers and handle NaNs
def preprocess_point_clouds(point_clouds):
    processed_point_clouds = []

    for pc in point_clouds:
        # Remove the header (first row)
        pc_cleaned = pc[2:]  # Skip the first row (['//X', 'Y', 'Z', 'intensity'])

        # Replace 'nan' values with 0 and convert the data to float
        pc_cleaned = pc_cleaned.astype(np.float32)  # Convert everything to float32

        processed_point_clouds.append(pc_cleaned)

    return processed_point_clouds

# Function to pad point clouds to have the same number of points
def pad_point_clouds(point_clouds, max_num_points):
    padded_point_clouds = []
    masks = []

    for pc in point_clouds:
        num_points = pc.shape[0]
        if num_points < max_num_points:
            # Pad with zeros
            pad_size = max_num_points - num_points
            padded_pc = np.pad(pc, ((0, pad_size), (0, 0)), 'constant', constant_values=0)  # Pad to the max_num_points
            mask = np.pad(np.ones(num_points), (0, pad_size), 'constant', constant_values=0)  # Create a mask
        else:
            padded_pc = pc[:max_num_points]  # Truncate if necessary
            mask = np.ones(max_num_points)  # Full mask if no padding is needed

        padded_point_clouds.append(padded_pc)
        masks.append(mask)

    return np.array(padded_point_clouds), np.array(masks)
```

```python
import torch
import torch.optim as optim

# Loss function and optimizer
criterion = torch.nn.MSELoss()  # Mean Squared Error for regression

# Preprocess the point clouds data
processed_point_clouds = preprocess_point_clouds(point_clouds)
test_processed_point_clouds = preprocess_point_clouds(test_point_clouds)

# Model instantiation
input_dim = 4  # Example: XYZ + intensity
output_dim = 1  # Predicting AGB or wood volume
model = PointNetRegression(input_dim=input_dim, output_dim=output_dim)

# Now, initialize the optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training loop
num_epochs = 25
batch_size = 8  # Adjust based on your GPU memory and dataset size

for epoch in range(num_epochs):
    model.train()

    total_loss = 0.0
    print(f"Starting epoch [{epoch+1}/{num_epochs}]...")

    for i in range(0, len(processed_point_clouds), batch_size):
        # Prepare batch data
        batch_point_clouds = processed_point_clouds[i:i+batch_size]
        batch_biomass_values = biomass_values[i:i+batch_size]

        # Find max number of points in the batch for padding
        max_num_points = max(pc.shape[0] for pc in batch_point_clouds)
        max_num_points = min(max_num_points, 10000)

        # Pad the point clouds and create masks
        padded_point_clouds, masks = pad_point_clouds(batch_point_clouds, max_num_points)

        # Convert to tensors
        point_clouds_tensor = torch.tensor(padded_point_clouds, dtype=torch.float32)  # Shape: (batch_size, max_num_points, input_dim)
        masks_tensor = torch.tensor(masks, dtype=torch.float32)  # Shape: (batch_size, max_num_points)
        biomass_values_tensor = torch.tensor(batch_biomass_values, dtype=torch.float32).unsqueeze(1)  # Shape: (batch_size, 1)

        optimizer.zero_grad()

        # print(point_clouds_tensor.shape)
        # print(masks_tensor.shape)
        # print(biomass_values_tensor.shape)

        # Forward pass with masking
        output = model(point_clouds_tensor, mask=masks_tensor)  # Pass the point clouds and mask through the model
        loss = criterion(output, biomass_values_tensor)  # Compute the loss

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    # Print average loss at the end of each epoch
    avg_loss = total_loss / len(processed_point_clouds)
    print(f'Epoch [{epoch+1}/{num_epochs}], Avg Loss: {avg_loss:.4f}')
```

```python
import numpy as np
from scipy.spatial import Delaunay

# Function to create a 3D graph from the point cloud using Delaunay triangulation
from sklearn.neighbors import NearestNeighbors

def create_knn_graph(points, k=6):
    # Create K-NN graph with k neighbors
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(points)
    distances, indices = nbrs.kneighbors(points)

    edges = set()
    for i in range(points.shape[0]):
        for j in indices[i]:
            edges.add((i, j))  # Add edge (i, j)
    return edges


# List to hold graphs for each point cloud
graphs = []
test_graphs=[]


for point_cloud in point_clouds:
    # Skip the header row if necessary (i.e., ['//X', 'Y', 'Z', 'intensity'])
    if isinstance(point_cloud[0][0], str):
        point_cloud = point_cloud[2:]  # Remove the header row

    # Convert point cloud to NumPy array for processing
    point_cloud_array = np.array(point_cloud, dtype=float)  # Ensure it is float

    # Create graph for current point cloud in 3D, handling NaN values
    try:
        edges = create_knn_graph(point_cloud_array, k=6)  # Using K=6 nearest neighbors
        graphs.append((point_cloud_array, edges))
  # Store points and their edges
    except ValueError as e:
        print(f"Skipping point cloud due to error: {e}")

# Now, 'graphs' contains a list of (point_cloud, edges) tuples, considering 3D space and NaN handling.


for point_cloud in test_point_clouds:
    # Skip the header row if necessary (i.e., ['//X', 'Y', 'Z', 'intensity'])
    if isinstance(point_cloud[0][0], str):
        point_cloud = point_cloud[2:]  # Remove the header row

    # Convert point cloud to NumPy array for processing
    test_point_cloud_array = np.array(point_cloud, dtype=float)  # Ensure it is float

    # Create graph for current point cloud in 3D, handling NaN values
    try:
        test_edges = create_knn_graph(test_point_cloud_array, k=6)  # Using K=6 nearest neighbors
        test_graphs.append((test_point_cloud_array, test_edges))
  # Store points and their edges
    except ValueError as e:
        print(f"Skipping point cloud due to error: {e}")
```
